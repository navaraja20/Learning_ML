```python
 import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/mall.csv')
```

```python
df.head()
```

```python
X=df.iloc[:,[3,4]].values
```

```python
from sklearn.cluster import MeanShift
ms=MeanShift(bandwidth=20)
ms.fit(X)
labels=ms.labels_
cluste_centers=ms.cluster_centers_
```

```python
n_clusters=len(np.unique(labels))
colors=['red','blue','green','yellow','purple','cyan','pink']
for i in range(n_clusters):
  cluster=X[labels==i]
  plt.scatter(cluster[:,0],cluster[:,1],s=100,c=colors[i],label=f'Cluster{i+1}')
plt.title('Mean Shift Clustering')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.legend()
plt.show()
```
